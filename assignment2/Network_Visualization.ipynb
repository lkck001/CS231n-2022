# Network Visualization

In this notebook, we will explore the use of *image gradients* for generating new images.

When training a model, we define a loss function which measures our current unhappiness with the model's performance. We then use backpropagation to compute the gradient of the loss with respect to the model parameters and perform gradient descent on the model parameters to minimize the loss.

Here we will do something slightly different. We will start from a CNN model which has been pretrained to perform image classification on the ImageNet dataset. We will use this model to define a loss function which quantifies our current unhappiness with our image. Then we will use backpropagation to compute the gradient of this loss with respect to the pixels of the image. We will then keep the model fixed and perform gradient descent *on the image* to synthesize a new image which minimizes the loss.

We will explore three techniques for image generation.

**Saliency Maps.** We can use saliency maps to tell which part of the image influenced the classification decision made by the network.

**Fooling Images.** We can perturb an input image so that it appears the same to humans but will be misclassified by the pretrained network.

**Class Visualization.** We can synthesize an image to maximize the classification score of a particular class; this can give us some sense of what the network is looking for when it classifies images of that class.

````python
import numpy as np
from scipy.fft import fft, ifft

def fit_trajectory_fourier(points, n_coefficients):
    """
    Fit an arbitrary trajectory using Fourier series
    
    Args:
        points: Array of shape (n_points, 3) containing x,y,z coordinates
        n_coefficients: Number of Fourier coefficients to use
        
    Returns:
        coefficients: Fourier coefficients for each dimension
        frequencies: Base frequencies
    """
    # Separate x, y, z coordinates
    x, y, z = points[:, 0], points[:, 1], points[:, 2]
    
    # Compute FFT for each dimension
    fx = fft(x)
    fy = fft(y)
    fz = fft(z)
    
    # Keep only the first n_coefficients
    fx[n_coefficients:] = 0
    fy[n_coefficients:] = 0
    fz[n_coefficients:] = 0
    
    return {
        'x_coeffs': fx,
        'y_coeffs': fy,
        'z_coeffs': fz
    }

def reconstruct_trajectory(coefficients, n_points):
    """
    Reconstruct trajectory from Fourier coefficients
    
    Args:
        coefficients: Dictionary containing Fourier coefficients
        n_points: Number of points to reconstruct
        
    Returns:
        Array of shape (n_points, 3) containing reconstructed trajectory
    """
    # Reconstruct using inverse FFT
    x = np.real(ifft(coefficients['x_coeffs']))[:n_points]
    y = np.real(ifft(coefficients['y_coeffs']))[:n_points]
    z = np.real(ifft(coefficients['z_coeffs']))[:n_points]
    
    return np.column_stack([x, y, z])

# Example usage
def example_with_arbitrary_path():
    # Generate some arbitrary 3D path
    t = np.linspace(0, 10, 1000)
    original_path = np.column_stack([
        3 * np.sin(t) + 0.5 * np.cos(5*t),  # Complex x movement
        2 * np.cos(t) - 0.3 * np.sin(3*t),  # Complex y movement
        0.5 * t * np.sin(t)                 # Complex z movement
    ])
    
    # Fit Fourier series
    coeffs = fit_trajectory_fourier(original_path, n_coefficients=50)
    
    # Reconstruct path
    reconstructed_path = reconstruct_trajectory(coeffs, len(t))
    
    # Plot original vs reconstructed
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D
    
    fig = plt.figure(figsize=(12, 5))
    
    # Original path
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot(original_path[:, 0], original_path[:, 1], original_path[:, 2])
    ax1.set_title('Original Path')
    
    # Reconstructed path
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.plot(reconstructed_path[:, 0], reconstructed_path[:, 1], reconstructed_path[:, 2])
    ax2.set_title('Reconstructed Path')
    
    plt.show()
    
    # Calculate error
    error = np.mean(np.sqrt(np.sum((original_path - reconstructed_path)**2, axis=1)))
    print(f"Average reconstruction error: {error:.6f}")

    return error

# Run example
example_with_arbitrary_path()
````

````python
import numpy as np
from scipy.fft import fft, ifft
from scipy.interpolate import interp1d

def enhanced_trajectory_fourier(points, n_coefficients=None, interpolation_points=None, error_threshold=1e-6):
    """
    Enhanced Fourier series fitting with adaptive coefficients and interpolation
    
    Args:
        points: Array of shape (n_points, 3) containing x,y,z coordinates
        n_coefficients: Initial number of coefficients (if None, will be determined adaptively)
        interpolation_points: Number of points to interpolate (if None, uses 2x original points)
        error_threshold: Maximum allowed error for adaptive coefficient selection
        
    Returns:
        coefficients: Fourier coefficients for each dimension
        frequencies: Base frequencies
        error: Approximation error
    """
    # Ensure uniform sampling through interpolation
    if interpolation_points is None:
        interpolation_points = len(points) * 2
    
    t_original = np.linspace(0, 1, len(points))
    t_interp = np.linspace(0, 1, interpolation_points)
    
    # Interpolate each dimension separately
    interpolators = [
        interp1d(t_original, points[:, i], kind='cubic', bounds_error=False, fill_value='extrapolate')
        for i in range(3)
    ]
    
    interpolated_points = np.column_stack([
        interp(t_interp) for interp in interpolators
    ])
    
    # Normalize the trajectory
    mean_pos = np.mean(interpolated_points, axis=0)
    std_pos = np.std(interpolated_points, axis=0)
    normalized_points = (interpolated_points - mean_pos) / std_pos
    
    def compute_fft_with_coeffs(points, n):
        """Compute FFT with specified number of coefficients"""
        fx = fft(points[:, 0])
        fy = fft(points[:, 1])
        fz = fft(points[:, 2])
        
        # Keep only n coefficients
        mask = np.zeros_like(fx, dtype=bool)
        mask[:n] = True
        fx[~mask] = 0
        fy[~mask] = 0
        fz[~mask] = 0
        
        return {'x_coeffs': fx, 'y_coeffs': fy, 'z_coeffs': fz}
    
    def reconstruct_and_error(coeffs, points):
        """Reconstruct trajectory and compute error"""
        reconstructed = reconstruct_trajectory(coeffs, len(points))
        error = np.mean(np.sqrt(np.sum((points - reconstructed)**2, axis=1)))
        return reconstructed, error
    
    # Adaptive coefficient selection if n_coefficients not specified
    if n_coefficients is None:
        n_coefficients = 10  # Start with 10 coefficients
        max_coeffs = len(points) // 2  # Maximum possible coefficients
        
        while n_coefficients < max_coeffs:
            coeffs = compute_fft_with_coeffs(normalized_points, n_coefficients)
            _, error = reconstruct_and_error(coeffs, normalized_points)
            
            if error < error_threshold:
                break
            
            n_coefficients *= 2
    
    # Compute final coefficients
    coeffs = compute_fft_with_coeffs(normalized_points, n_coefficients)
    
    # Store normalization parameters for reconstruction
    coeffs['mean'] = mean_pos
    coeffs['std'] = std_pos
    
    # Compute final error
    _, final_error = reconstruct_and_error(coeffs, normalized_points)
    
    return coeffs, final_error

def enhanced_reconstruct_trajectory(coefficients, n_points):
    """
    Enhanced reconstruction with denormalization
    
    Args:
        coefficients: Dictionary containing Fourier coefficients and normalization params
        n_points: Number of points to reconstruct
        
    Returns:
        Array of shape (n_points, 3) containing reconstructed trajectory
    """
    # Reconstruct normalized trajectory
    x = np.real(ifft(coefficients['x_coeffs']))[:n_points]
    y = np.real(ifft(coefficients['y_coeffs']))[:n_points]
    z = np.real(ifft(coefficients['z_coeffs']))[:n_points]
    
    reconstructed = np.column_stack([x, y, z])
    
    # Denormalize
    reconstructed = (reconstructed * coefficients['std']) + coefficients['mean']
    
    return reconstructed

def analyze_trajectory_complexity(points):
    """
    Analyze trajectory complexity to suggest optimal parameters
    
    Args:
        points: Array of shape (n_points, 3)
    
    Returns:
        dict with suggested parameters
    """
    # Compute velocity and acceleration
    velocity = np.diff(points, axis=0)
    acceleration = np.diff(velocity, axis=0)
    
    # Compute complexity metrics
    curvature = np.mean(np.sqrt(np.sum(acceleration**2, axis=1)))
    
    # Suggest parameters based on complexity
    suggested_coeffs = int(max(10, min(len(points) // 2, curvature * 20)))
    suggested_interp_points = len(points) * 2
    
    return {
        'suggested_coefficients': suggested_coeffs,
        'suggested_interp_points': suggested_interp_points,
        'curvature': curvature
    }

# Example usage with error analysis
def example_with_precision_analysis():
    # Generate complex test trajectory
    t = np.linspace(0, 10, 1000)
    original_path = np.column_stack([
        3 * np.sin(t) + 0.5 * np.cos(5*t) + 0.2 * np.sin(10*t),
        2 * np.cos(t) - 0.3 * np.sin(3*t) + 0.15 * np.cos(7*t),
        0.5 * t * np.sin(t) + 0.1 * np.cos(4*t)
    ])
    
    # Analyze trajectory complexity
    analysis = analyze_trajectory_complexity(original_path)
    
    # Fit with suggested parameters
    coeffs, error = enhanced_trajectory_fourier(
        original_path,
        n_coefficients=analysis['suggested_coefficients'],
        interpolation_points=analysis['suggested_interp_points']
    )
    
    # Reconstruct
    reconstructed_path = enhanced_reconstruct_trajectory(coeffs, len(t))
    
    # Compute detailed error metrics
    point_errors = np.sqrt(np.sum((original_path - reconstructed_path)**2, axis=1))
    max_error = np.max(point_errors)
    mean_error = np.mean(point_errors)
    std_error = np.std(point_errors)
    
    print(f"Trajectory Analysis:")
    print(f"Curvature: {analysis['curvature']:.6f}")
    print(f"Suggested coefficients: {analysis['suggested_coefficients']}")
    print(f"\nError Metrics:")
    print(f"Mean Error: {mean_error:.6f}")
    print(f"Max Error: {max_error:.6f}")
    print(f"Error Std: {std_error:.6f}")
    
    return coeffs, error, reconstructed_path
example_with_precision_analysis()
````

# Saliency Maps
Using this pretrained model, we will compute class saliency maps as described in Section 3.1 of [2].

A **saliency map** tells us the degree to which each pixel in the image affects the classification score for that image. To compute it, we compute the gradient of the unnormalized score corresponding to the correct class (which is a scalar) with respect to the pixels of the image. If the image has shape `(3, H, W)` then this gradient will also have shape `(3, H, W)`; for each pixel in the image, this gradient tells us the amount by which the classification score will change if the pixel changes by a small amount. To compute the saliency map, we take the absolute value of this gradient, then take the maximum value over the 3 input channels; the final saliency map thus has shape `(H, W)` and all entries are nonnegative.

[2] Karen Simonyan, Andrea Vedaldi, and Andrew Zisserman. "Deep Inside Convolutional Networks: Visualising
Image Classification Models and Saliency Maps", ICLR Workshop 2014.

`s.gather(1, y.view(-1, 1)).squeeze()`

will be a PyTorch Tensor of shape `(N,)` containing one entry from each row of `s`, selected according to the indices in `y`.

# Class Visualization
By starting with a random noise image and performing gradient ascent on a target class, we can generate an image that the network will recognize as the target class. This idea was first presented in [2]; [3] extended this idea by suggesting several regularization techniques that can improve the quality of the generated image.

Concretely, let $I$ be an image and let $y$ be a target class. Let $s_y(I)$ be the score that a convolutional network assigns to the image $I$ for class $y$; note that these are raw unnormalized scores, not class probabilities. We wish to generate an image $I^*$ that achieves a high score for the class $y$ by solving the problem

$$
I^* = \arg\max_I (s_y(I) - R(I))
$$

where $R$ is a (possibly implicit) regularizer (note the sign of $R(I)$ in the argmax: we want to minimize this regularization term). We can solve this optimization problem using gradient ascent, computing gradients with respect to the generated image. We will use (explicit) L2 regularization of the form

$$
R(I) = \lambda \|I\|_2^2
$$

**and** implicit regularization as suggested by [3] by periodically blurring the generated image. We can solve this problem using gradient ascent on the generated image.

[2] Karen Simonyan, Andrea Vedaldi, and Andrew Zisserman. "Deep Inside Convolutional Networks: Visualising
Image Classification Models and Saliency Maps", ICLR Workshop 2014.

[3] Yosinski et al, "Understanding Neural Networks Through Deep Visualization", ICML 2015 Deep Learning Workshop

````python
def create_class_visualization(target_y, model, dtype, **kwargs):
    """
    Generate an image to maximize the score of target_y under a pretrained model.

    Inputs:
    - target_y: Integer in the range [0, 1000) giving the index of the class
    - model: A pretrained CNN that will be used to generate the image
    - dtype: Torch datatype to use for computations

    Keyword arguments:
    - l2_reg: Strength of L2 regularization on the image
    - learning_rate: How big of a step to take
    - num_iterations: How many iterations to use
    - blur_every: How often to blur the image as an implicit regularizer
    - max_jitter: How much to gjitter the image as an implicit regularizer
    - show_every: How often to show the intermediate result
    """
    model.type(dtype)
    l2_reg = kwargs.pop('l2_reg', 1e-3)
    learning_rate = kwargs.pop('learning_rate', 25)
    num_iterations = kwargs.pop('num_iterations', 100)
    blur_every = kwargs.pop('blur_every', 10)
    max_jitter = kwargs.pop('max_jitter', 16)
    show_every = kwargs.pop('show_every', 25)

    # Randomly initialize the image as a PyTorch Tensor, and make it requires gradient.
    img = torch.randn(1, 3, 224, 224).mul_(1.0).type(dtype).requires_grad_()

    for t in range(num_iterations):
        # Randomly jitter the image a bit; this gives slightly nicer results
        ox, oy = random.randint(0, max_jitter), random.randint(0, max_jitter)
        img.data.copy_(jitter(img.data, ox, oy))
        class_visualization_update_step(img, model, target_y, l2_reg, learning_rate)
        # Undo the random jitter
        img.data.copy_(jitter(img.data, -ox, -oy))

        # As regularizer, clamp and periodically blur the image
        for c in range(3):
            lo = float(-SQUEEZENET_MEAN[c] / SQUEEZENET_STD[c])
            hi = float((1.0 - SQUEEZENET_MEAN[c]) / SQUEEZENET_STD[c])
            img.data[:, c].clamp_(min=lo, max=hi)
        if t % blur_every == 0:
            blur_image(img.data, sigma=0.5)

        # Periodically show the image
        if t == 0 or (t + 1) % show_every == 0 or t == num_iterations - 1:
            plt.imshow(deprocess(img.data.clone().cpu()))
            class_name = class_names[target_y]
            plt.title('%s\nIteration %d / %d' % (class_name, t + 1, num_iterations))
            plt.gcf().set_size_inches(4, 4)
            plt.axis('off')
            plt.show()

    return deprocess(img.data.cpu())
````
