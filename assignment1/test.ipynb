```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import rgb_to_hsv, hsv_to_rgb

def visualize_rgb_hsv_conversion():
    # Create sample RGB image
    rgb_image = np.zeros((100, 300, 3))
    print("RGB image shape:", rgb_image.shape)  # Will show (100, 300, 3)
    # Create three colored regions: Red, Green, Blue
    rgb_image[:, 0:100] = [1, 0, 0]    # Red
    rgb_image[:, 100:200] = [0, 1, 0]  # Green
    rgb_image[:, 200:300] = [0, 0, 1]  # Blue
    
    # Convert to HSV
    hsv_image = rgb_to_hsv(rgb_image)
    
    # Extract H, S, V channels
    h = hsv_image[:, :, 0]  # Hue
    s = hsv_image[:, :, 1]  # Saturation
    v = hsv_image[:, :, 2]  # Value
    
    # Visualize
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # Original RGB
    axes[0,0].imshow(rgb_image)
    axes[0,0].set_title('Original RGB')
    
    # HSV channels
    axes[0,1].imshow(h, cmap='hsv')
    axes[0,1].set_title('Hue Channel')
    
    axes[0,2].imshow(s, cmap='gray')
    axes[0,2].set_title('Saturation Channel')
    
    axes[1,0].imshow(v, cmap='gray')
    axes[1,0].set_title('Value Channel')
    
    # Print values
    print("RGB Values:")
    print("Red:   ", rgb_image[0,0])    # [1, 0, 0]
    print("Green: ", rgb_image[0,100])  # [0, 1, 0]
    print("Blue:  ", rgb_image[0,200])  # [0, 0, 1]
    
    print("\nHSV Values:")
    print("Red:   ", hsv_image[0,0])    # [0, 1, 1]
    print("Green: ", hsv_image[0,100])  # [0.33, 1, 1]
    print("Blue:  ", hsv_image[0,200])  # [0.67, 1, 1]
    
    plt.show()
visualize_rgb_hsv_conversion()
# Create example with color gradient
def create_color_wheel():
    size = 100
    x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))
    
    # Convert to polar coordinates
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    
    # Create HSV image
    hsv = np.zeros((size, size, 3))
    hsv[:,:,0] = (theta + np.pi) / (2*np.pi)  # Hue
    hsv[:,:,1] = np.clip(r, 0, 1)             # Saturation
    hsv[:,:,2] = 1                            # Value
    
    # Convert to RGB
    rgb = hsv_to_rgb(hsv)
    
    return rgb, hsv

# Display results
rgb_wheel, hsv_wheel = create_color_wheel()
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(rgb_wheel)
plt.title('RGB Color Wheel')

plt.subplot(132)
plt.imshow(hsv_wheel[:,:,0], cmap='hsv')
plt.title('Hue Channel')

plt.subplot(133)
plt.imshow(hsv_wheel[:,:,1], cmap='gray')
plt.title('Saturation Channel')

plt.show()
```
